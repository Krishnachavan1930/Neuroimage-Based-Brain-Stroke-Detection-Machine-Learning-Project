```python
# import the necessary packages
from tensorflow.keras.models import load_model
from collections import deque
import numpy as np
import argparse
import pickle
import cv2
```

```python
# load the trained model and label binarizer from disk
print("[INFO] loading model and label binarizer...")
model = load_model('model-cnn.h5')
lb = pickle.loads(open('labels-cnn.h5', "rb").read())
```

```python
image = cv2.imread('./test_images/stroke (1).jpg', 0)
image = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)
image = cv2.resize (image, (224,224))
image = np.resize(image,(1,image.shape[0],image.shape[1],image.shape[2]))
image = image/255.

pred = model.predict(image)

print(lb.inverse_transform(pred))
```
