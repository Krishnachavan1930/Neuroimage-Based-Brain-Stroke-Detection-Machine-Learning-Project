# Brain stroke prediction using CNN-Transfer Learning

### import necessary packages

```python
import os
import cv2
import numpy as np
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from sklearn.preprocessing import LabelBinarizer
from sklearn.model_selection import train_test_split
from keras.utils import Sequence
from keras.layers import Dense, Flatten, Conv2D, MaxPooling2D, BatchNormalization, Dropout
from keras.layers import Dense, Flatten, Dropout
from keras.models import Model, Sequential
from keras.optimizers import Adam, Nadam, SGD
import matplotlib.pyplot as plt
import pickle
```

### importing Dataset

```python
datasetpath = './Dataset'

# Global Constants
BS = 64
LR = 1e-2
EPOCHS = 25
```

```python
print('[INFO Extracting Data and Labels')


def data_labels(datasetpath):
    data = []
    labels = []
    for dirname in os.listdir(datasetpath):
        dirpath = os.path.join(datasetpath, dirname)
        label = dirname.split('.')[-1]
        for filename in os.listdir(dirpath):
            filepath = os.path.join(dirpath, filename)
            print(filepath)
            image = cv2.imread(filepath, 1)
            image = cv2.resize(image, (224, 224))
            data.append(image)
            labels.append(label)
    return np.array(data, dtype='float32') / 255., np.array(labels)


data, labels = data_labels(datasetpath)
```

### Label Binarization

```python
lb = LabelBinarizer()
labels = lb.fit_transform(labels)
```

### Split dataset

```python
xTrain, xTest, yTrain, yTest = train_test_split(data, labels, test_size=0.25, stratify=labels, random_state=42)
xTrain.shape, yTrain.shape
```

### # Define the Network

```python
# Model building
activation = 'relu'

model = Sequential()
model.add(Conv2D(16, 5, activation = activation, padding = 'same', input_shape = (224,224,3)))
model.add(MaxPooling2D())
model.add(Dropout(0.2))
model.add(Conv2D(16, 3, activation = activation, padding = 'same', kernel_initializer = 'he_uniform'))
model.add(MaxPooling2D())
model.add(Dropout(0.2))
model.add(Conv2D(32, 5, activation = activation, padding = 'same', kernel_initializer = 'he_uniform'))
model.add(MaxPooling2D())
model.add(Dropout(0.2))
model.add(Conv2D(32, 2, activation = activation, padding = 'same', kernel_initializer = 'he_uniform'))
model.add(MaxPooling2D())
model.add(Dropout(0.5))
model.add(Flatten())
x = model.output  
x = Dense(32, activation = activation, kernel_initializer = 'he_uniform')(x)
prediction_layer = Dense(1, activation = 'sigmoid')(x)

# Make a new model combining both feature extractor and x
cnn_model = Model(inputs=model.input, outputs=prediction_layer)
cnn_model.compile(optimizer='rmsprop',loss = 'binary_crossentropy', metrics = ['accuracy'])
print(cnn_model.summary()) 

```

### Training Data

```python
print('[INFO] Training Started')
H = cnn_model.fit(
        xTrain,yTrain,
        steps_per_epoch=len(xTrain)//BS,
        epochs = EPOCHS,
        validation_data=(xTest,yTest),
        validation_steps=len(xTest)//BS,
         )
```

### Plot

```python
N = EPOCHS
plt.style.use("ggplot")
plt.figure()
plt.plot(np.arange(0, N), H.history["loss"], label="train_loss")
plt.plot(np.arange(0, N), H.history["val_loss"], label="val_loss")
plt.plot(np.arange(0, N), H.history["accuracy"], label="train_acc")
plt.plot(np.arange(0, N), H.history["val_accuracy"], label="val_acc")
plt.title("Training Loss and Accuracy on Dataset")
plt.xlabel("Epoch #")
plt.ylabel("Loss/Accuracy")
plt.legend(loc="lower right")
plt.savefig('./graphs/graph-cnn.png')
```

### Saving Model to disk

```python
print("[INFO] serializing network...")
cnn_model.save('model-cnn.h5', save_format="h5")
# serialize the label binarizer to disk
f = open('labels-cnn.h5', "wb")
f.write(pickle.dumps(lb))
f.close()
```

```python
from tensorflow.keras.models import load_model
# load the trained model from disk
print("[INFO] loading model")
model = load_model('model-cnn.h5')
```

### Evaluate Model

```python
ypred = model.predict(xTest)
ypred = lb.inverse_transform(ypred)
```

```python
yTest = lb.inverse_transform(yTest)
```

#### Confusion Matrix

```python
from sklearn.metrics import confusion_matrix, classification_report
```

#### classification report
